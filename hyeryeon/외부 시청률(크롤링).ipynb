```python
from bs4 import BeautifulSoup
import requests
from datetime import datetime, timedelta
from dateutil.relativedelta import  relativedelta
from tqdm import tqdm_notebook
import pandas as pd
```

```python
date = pd.date_range(start='2019-01-01', end='2019-12-31')
```

```python
temp_지상파 = []
날짜 = []

startdate = 0
for i in tqdm_notebook(date):
    
    ds = str(i).split(' ')[0].replace('-', '')
    
    url = 'http://www.nielsenkorea.co.kr/tv_terrestrial_day.asp?menu=Tit_1&sub_menu=1_1&area=01&begin_date={}'.format(ds)
    날짜.append(ds)
    
    result = requests.get(url)
    soup = BeautifulSoup(result.text, 'html.parser')
    tbody = soup.select('table.ranking_tb td.percent')

    rate_sum = 0
    for i in tbody[:10]:
        rate = i.text.strip().replace(',', '.')
        rate_sum += float(rate)
    temp_지상파.append(rate_sum)
```

```python
temp_케이블 = []
날짜 = []

startdate = 0
for i in tqdm_notebook(date):
    
    ds = str(i).split(' ')[0].replace('-', '')
    
    url = 'https://www.nielsenkorea.co.kr/tv_terrestrial_day.asp?menu=Tit_1&sub_menu=3_1&area=01&begin_date={}'.format(ds)

    result = requests.get(url)
    soup = BeautifulSoup(result.text, 'html.parser')
    tbody = soup.select('table.ranking_tb td.percent')

    rate_sum = 0
    for i in tbody[:10]:
        rate = i.text.strip().replace(',', '.')
        rate_sum += float(rate)
    temp_케이블.append(rate_sum)
```

```python
temp_종편 = []
날짜 = []

startdate = 0
for i in tqdm_notebook(date):
    
    ds = str(i).split(' ')[0].replace('-', '')
    
    url = 'http://www.nielsenkorea.co.kr/tv_terrestrial_day.asp?menu=Tit_1&sub_menu=2_1&area=01&begin_date={}'.format(ds)

    result = requests.get(url)
    soup = BeautifulSoup(result.text, 'html.parser')
    tbody = soup.select('table.ranking_tb td.percent')

    rate_sum = 0
    for i in tbody[:10]:
        rate = i.text.strip().replace(',', '.')
        rate_sum += float(rate)
    temp_종편.append(rate_sum)
```
