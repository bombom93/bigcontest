```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from datetime import datetime
```

```python
dt_index=pd.date_range(start='20180101',end='20200630')
```

```python
dt_list=dt_index.strftime('%Y%m%d').tolist()
```
